```python
import matplotlib.pyplot as plt

x = [1800, 1600, 1400, 1200, 1000, 800, 600, 400, 200][::-1]
y = [2.02, 1.9, 1.6, 1.7, 1.6, 1.5, 1.6, 0.5, 0.0][::-1]
fig = plt.figure(figsize=(10, 6))
plt.plot(x,y, 'o-')
fig.suptitle('BLEU SCORE FOR TEST DATA')
plt.ylabel('BLEU')
plt.xlabel('train steps')
plt.show()
```
